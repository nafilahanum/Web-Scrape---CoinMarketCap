```python
# Import Library
import pandas as pd
import requests
from bs4 import BeautifulSoup

```

```python
# List kosong
nama_kripto_list = []
market_cap_list = []
price_list = []
circulating_supply_list = []
symbol_list = []
```

```python

# Function untuk scrape data
def scrape(date='20240825/'):
    # URL website
    url = 'https://coinmarketcap.com/historical/' + date

    # Requests website
    webpage = requests.get(url)
    # Parse text dari web
    soup = BeautifulSoup(webpage.text, 'html.parser')

    # Ambil elemen tabel kolom
    table_row = soup.find_all('tr', attrs={'class': 'cmc-table-row'})

    count = 0

    for row in table_row:
        if count == 11:  # Ambil hanya 10 data
            break
        count += 1

        # Cari kolom
        nama_kolom = row.find('td', attrs={'class': 'cmc-table__cell--sort-by__name'})
        nama_kripto = nama_kolom.find('a', attrs={'class': 'cmc-link'}).text.strip()

        market_cap = row.find('td', attrs={'class': 'cmc-table__cell--sort-by__market-cap'}).text.strip()
        price = row.find('td', attrs={'class': 'cmc-table__cell--sort-by__price'}).text.strip()
        circulating_supply_symbol = row.find('td', attrs={'class': 'cmc-table__cell--sort-by__circulating-supply'}).text.strip()

        # Split data
        circulating_supply = circulating_supply_symbol.split(' ')[0]
        symbol = circulating_supply_symbol.split(' ')[1]

        # Tambah data ke list
        nama_kripto_list.append(nama_kripto)
        market_cap_list.append(market_cap)
        price_list.append(price)
        circulating_supply_list.append(circulating_supply)
        symbol_list.append(symbol)

    # Buat DataFrame dari list
    df = pd.DataFrame({
        'Nama': nama_kripto_list,
        'Market Cap': market_cap_list,
        'Price': price_list,
        'Circulating Supply': circulating_supply_list,
        'Symbol': symbol_list
    })

    return df

```
